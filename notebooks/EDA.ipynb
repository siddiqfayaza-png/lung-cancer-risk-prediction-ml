Encoding

```python
# 1. Separate categorical columns
cat_cols = df.select_dtypes(include="str").columns

# 2. Encode
df_encoded = pd.get_dummies(df, columns=cat_cols)

# 3. Correlation
corr = df_encoded.corr()

# 4. Heatmap
sns.heatmap(corr)
```
